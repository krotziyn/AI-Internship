```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L12-v2')
sentences = ["Yapay zeka bulutta çalışıyor.", "Staj projem harika gidiyor."]
embeddings = model.encode(sentences)

for sentence, embedding in zip(sentences, embeddings):
    print(f"Cümle: {sentence}")
    print(f"Vektörün ilk 5 sayısı: {embedding[:5]}")
```

```python
from sentence_transformers import util

# Karşılaştırmak istediğin iki cümleyi seç
cumle1 = model.encode("Yapay zeka projeleri çok heyecan verici.")
cumle2 = model.encode("Makine öğrenmesi çalışmaları oldukça ilginç.")

# Benzerliği hesapla (0 ile 1 arası bir sonuç verir)
benzerlik = util.cos_sim(cumle1, cumle2)

print(f"Bu iki cümlenin birbirine benzerlik oranı: %{benzerlik.item() * 100:.2f}")
```
