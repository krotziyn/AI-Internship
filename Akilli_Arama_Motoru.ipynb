```python
from sentence_transformers import SentenceTransformer, util
import torch

# Load the model
model = SentenceTransformer('all-MiniLM-L12-v2')

# Our English Knowledge Base
knowledge_base = [
    "You can delete unnecessary files to increase computer performance.",
    "Python is a very popular programming language in data science and AI.",
    "It is recommended to drink at least 2 liters of water daily for a healthy life.",
    "Github is a platform where developers upload, store and share their codes.",
    "AI models are trained using large datasets."
]

# Encode sentences to vectors
knowledge_embeddings = model.encode(knowledge_base, convert_to_tensor=True)

print("System is ready with English Knowledge Base!")
```

```python
def smart_search(query):
    # Convert query to vector
    query_embedding = model.encode(query, convert_to_tensor=True)

    # Calculate similarity scores
    scores = util.cos_sim(query_embedding, knowledge_embeddings)[0]

    # Get top 3 results
    top_results = torch.topk(scores, k=3)

    print(f"\n--- SEARCH RESULTS FOR: '{query}' ---")
    for i in range(len(top_results[0])):
        score = top_results[0][i].item()
        index = top_results[1][i].item()
        print(f"{i+1}. Result: {knowledge_base[index]} (Score: {score:.4f})")
```

```python
print("--- STARTING THE SMART SEARCH ENGINE ---")
print("(Type 'exit' to quit)\n")

while True:
    user_input = input("Enter your search query: ")

    if user_input.lower() == 'exit':
        print("Closing the system. Goodbye!")
        break

    if user_input.strip() == "":
        continue

    smart_search(user_input)
```
