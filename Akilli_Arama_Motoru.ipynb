```python
from sentence_transformers import SentenceTransformer, util
import torch

# Modeli yüklüyoruz (İngilizce odaklı hızlı model)
model = SentenceTransformer('all-MiniLM-L12-v2')

# İngilizce Bilgi Bankamız
bilgi_bankasi = [
    "You can delete unnecessary files to increase computer performance.",
    "Python is a very popular programming language in data science and AI.",
    "It is recommended to drink at least 2 liters of water daily for a healthy life.",
    "Github is a platform where developers upload, store and share their codes.",
    "AI models are trained using large datasets."
]

# Cümleleri vektörlere çeviriyoruz
bilgi_vektorleri = model.encode(bilgi_bankasi, convert_to_tensor=True)

print("English Knowledge Base is ready!")
```

```python
def ara(sorgu):
    # 1. Sorguyu vektöre çevir
    sorgu_vektoru = model.encode(sorgu, convert_to_tensor=True)

    # 2. Benzerlik skorlarını hesapla
    skorlar = util.cos_sim(sorgu_vektoru, bilgi_vektorleri)[0]

    # 3. En iyi 3 sonucu puanlarıyla getir
    top_results = torch.topk(skorlar, k=3)

    print(f"\n--- SEARCH QUERY: '{sorgu}' ---")
    for i in range(len(top_results[0])):
        skor = top_results[0][i].item()
        index = top_results[1][i].item()
        print(f"{i+1}. Result: {bilgi_bankasi[index]} (Score: {skor:.4f})")

# TEST EDELİM
ara("where can I upload my codes")
ara("how to make my pc faster")
ara("importance of drinking water")
```
